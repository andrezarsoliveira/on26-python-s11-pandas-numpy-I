# Introdução


Anime é um estilo de animação originário do Japão que abrange uma ampla variedade de gêneros e estilos. A palavra "anime" é uma abreviação da palavra inglesa "animation" e, no Japão, refere-se a qualquer tipo de animação, seja ela japonesa ou não. No entanto, fora do Japão, o termo é comumente usado para se referir especificamente às animações produzidas no Japão ou influenciadas pelo estilo japonês.

O anime abrange uma gama diversificada de temas, desde aventuras épicas e ficção científica até romances, comédias e histórias cotidianas. Os personagens frequentemente apresentam características distintas, como olhos grandes e expressivos, estilos de cabelo únicos e traços faciais distintos. As narrativas podem ser complexas e envolventes, explorando temas profundos e questões filosóficas, ou podem ser leves e destinadas ao entretenimento puro.

Uma característica marcante do anime é a capacidade de atrair públicos de todas as idades. Existem animes voltados para crianças, adolescentes e adultos, e muitos deles conquistam fãs ao redor do mundo. Alguns animes também são adaptados de mangás, que são histórias em quadrinhos japonesas.

Ao longo dos anos, o anime tornou-se uma parte significativa da cultura pop global, influenciando a indústria do entretenimento em diversos países. A comunidade de fãs de anime é vasta e apaixonada, participando de convenções, eventos e compartilhando sua paixão online.

```python
import pandas as pd
```

```python
df_anime = pd.read_csv ('/content/Anime.csv')
```

```python
# criando a função de processamento dos dados
def processamento_df(df_anime):

    # 3 primeiras linhas do df
    df_3_linhas = df_anime.head(3)

    # selecionando aleatoriamente 100 linhas do df
    df_menor = df_anime.sample(100)

    # removendo colunas com muitos dados nulos
    df_anime = df_anime.drop(columns='staff')

    # ordenando por conteúdo sensível
    df_anime = df_anime.sort_values("Content_Warning")

    # removendo dados duplicados
    df_anime = df_anime.drop_duplicates()

    # reset index
    df_anime = df_anime.reset_index()

    # lista de colunas
    colunas = list(df_anime.columns)

    # metricas: analise descritiva
    metricas = df_anime.describe(include='all')

    return df_3_linhas, df_menor, df_anime, colunas, metricas


```

```python
 # chamando a função processamento usando os dados do df
df_3_linhas, df_menor, df_teste, colunas, metricas = processamento_df(df_anime)
```
